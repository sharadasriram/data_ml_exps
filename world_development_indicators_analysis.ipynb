```python
country_codes = df_wb['country_code'].unique()
null_cols_to_fillna = ['female_emp_salaried_%',
        'food_insecurity_popln_%', 'maternal_mort_rate', 'poverty_gap_%',
        'poverty_popln_%', 'prim_completion_%', 'sec_completion_%',
        'total_fert_rate', 'under_5_mort_rate', 'women_parliment_%']
master_df = pd.DataFrame()
master_lr_dict = {}
for cc in country_codes:
    df = pd.read_csv(f"data_wb_{cc}.csv")
    master_lr_dict[cc] = {}
    for col in null_cols_to_fillna:
        df_new = df[[col, 'year']].dropna()
        if df_new.shape[0] > 1:
            lr = LinearRegression()
            lr.fit(df_new[['year']], df_new[col])
            df[col].fillna(pd.Series(lr.predict(df[['year']])), inplace=True)
            master_lr_dict[cc][col] = lr
    master_df = pd.concat([master_df, df])
    print(master_df['country_name'].unique())
```

```python
d1 = {}
for idx, row in master_df[['country_code', 'country_name']].drop_duplicates().iterrows():
    d1[row['country_code']] = row['country_name']
d1
```

```python
d = {'country_name':[], 'country_code':[], 'year':[], 'female_emp_salaried_%':[], 'food_insecurity_popln_%':[], 
     'maternal_mort_rate':[], 'poverty_gap_%':[], 'poverty_popln_%':[], 'prim_completion_%':[], 'sec_completion_%':[], 
     'total_fert_rate':[], 'under_5_mort_rate':[], 'women_parliment_%':[]}
for country_code, val in master_lr_dict.items():
    d['country_name'].append(d1[country_code])
    d['country_code'].append(country_code)
    d['year'].append(2030)
    for col, lr in val.items():
        pred_2030 = lr.predict([[2030]])[0]
        if pred_2030 < 0:
            pred_2030 = 0
        d[col].append(pred_2030)
```
